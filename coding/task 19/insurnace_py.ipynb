```python
import numpy as np                  #importing all the needed modules
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("insurance.csv")          # making our data frame
df.head()
```

```python
x = df.iloc[:,:1].values   #extracts values from the first column and assigns them to x
y = df.iloc[:,-1].values   # extracts all the values from the last line and assigns them to y


plt.scatter(x,y,color = 'b') # plotting a scatterplot
plt.xlabel('age')
plt.ylabel('charges')
plt.show()
```

```python
charges_model = LinearRegression()  # using linear regression to estimate the relationship between age and charges
charges_model.fit(x,y)              # used to calculate the coefficients
```

```python
y_pred = charges_model.predict(x)    # generating predicted values

plt.scatter(x,y,color = 'b')           # generating new scatterplot with line of best fit 
plt.plot(x,y_pred,color = 'r')
plt.xlabel('age')
plt.ylabel('charges')
plt.show()
```

```python
unkown_x = [[70]]           # declaring an unknown value of x

x_pred = np.append(x, unkown_x).reshape(-1,1)
y_pred = charges_model.predict(x_pred)

plt.scatter(x,y,color = 'b')
plt.plot(x_pred,y_pred,color = 'r')
plt.xlabel('age')
plt.ylabel('charges')
plt.show()

print( charges_model.predict(unkown_x))
```
